```python
import torch
import sklearn
import numpy as np
from sklearn.manifold import TSNE
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def preprocess_features(fake_features, real_features, N):
    subsample_idc_fake = np.random.choice(fake_features.shape[0], N, replace=False)
    fake_features_500 = fake_features[subsample_idc_fake,:]
    subsample_idc_real = np.random.choice(real_features.shape[0], N, replace=False)
    real_features_500 = real_features[subsample_idc_real,:]

    labels_fake = np.ones(N)
    labels_real = np.zeros(N)

    features = np.concatenate((fake_features_500, real_features_500))
    labels = np.concatenate((labels_fake,labels_real))

    print(features.shape)
    print(labels.shape)
    return features, labels

def get_tsne(features):
    n_components = 2
    tsne = TSNE(n_components=2, verbose=1, perplexity=30, n_iter=500)
    tsne_result = tsne.fit_transform(features)
    return tsne_result

def plot_tsne(tsne_result,labels):
    tsne_result_df = pd.DataFrame({'tsne_1': tsne_result[:,0], 'tsne_2': tsne_result[:,1], 'label': labels})
    fig, ax = plt.subplots(1)
    fig.set_size_inches(18.5, 10.5)
    sns.scatterplot(x='tsne_1', y='tsne_2', hue='label', data=tsne_result_df, ax=ax,s=120, size=10, legend=False)
    lim = (tsne_result.min()-5, tsne_result.max()+5)
    ax.set_xlim(lim)
    ax.set_ylim(lim)
    ax.set_aspect('equal')
    ax.set_axis_off()
    ax.legend(loc='upper left')

def pipeline(fake_features, real_features, N):
    features, labels = preprocess_features(fake_features, real_features, N)
    tsne_result = get_tsne(features)
    plot_tsne(tsne_result, labels)
```

```python
fake_features = torch.load("features/vitL14_datacomp_progan_fake_features.pt")
real_features = torch.load("features/vitL14_datacomp_progan_real_features.pt")
pipeline(fake_features, real_features,5000)
```
